```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
breast = load_breast_cancer()
breast_data = breast.data
breast_data.shape
```

```python
breast_labels = breast.target
breast_labels.shape
```

```python
labels = np.reshape(breast_labels,(569,1))
```

```python
final_breast_data = np.concatenate([breast_data,labels],axis=1)
final_breast_data.shape
```

```python
breast_dataset = pd.DataFrame(final_breast_data)
features = breast.feature_names
features
```

```python
features_labels = np.append(features,'label')
breast_dataset.columns = features_labels
breast_dataset.head()
```

```python
breast_dataset['label'].replace(0, 'Benign',inplace=True)
breast_dataset['label'].replace(1, 'Malignant',inplace=True)
breast_dataset.tail()
```

```python
#veri gÃ¶rselleÅŸtirme
from sklearn.preprocessing import StandardScaler
x = breast_dataset.loc[:, features].values
x = StandardScaler().fit_transform(x) # normalizing the features
```

```python
x.shape
```

```python
np.mean(x),np.std(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_breast = pd.DataFrame(x,columns=feat_cols)
normalised_breast.tail()
```

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
```

```python
principal_breast_Df = pd.DataFrame(data = principalComponents_breast
             , columns = ['PCA-1', 'PCA-2'])
principal_breast_Df.tail()
```

```python
print('Varyans {}'.format(pca_breast.explained_variance_ratio_))
```

```python

%matplotlib inline


plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('PCA-1',fontsize=20)
plt.ylabel('PCA-2',fontsize=20)
plt.title("PCA Breast Cancer Veri Seti",fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = breast_dataset['label'] == target
    plt.scatter(principal_breast_Df.loc[indicesToKeep, 'PCA-1']
               , principal_breast_Df.loc[indicesToKeep, 'PCA-2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```
