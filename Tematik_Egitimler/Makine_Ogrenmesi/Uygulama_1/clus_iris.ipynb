```python
data_out = data['Species'].replace(['Iris-setosa', 
                                    'Iris-versicolor', 
                                    'Iris-virginica'],
                                   [0, 1, 2],
                                   inplace=False)
```

```python
data_list = []

for index, row in data.iterrows():
    data_list.append([row['SepalLengthCm'],
                       row['SepalWidthCm'],
                       row['PetalLengthCm'],
                       row['PetalWidthCm'], 
                      data_out[index]])
```

```python
if shuffle:
    for x in range(5):
        random.shuffle(data_list)

data_list
```

```python
split_point = int((len(data_list) / k) * (k - 1))

data_train, data_test = data_list[:split_point], data_list[split_point:]
```

```python
from sklearn.model_selection import KFold

splitter = KFold(n_splits=k)

split_indexes = []
for fold_id, (train_index, test_index) in enumerate(splitter.split(data_list)):
    print("Train" + str(train_index))
    print("Test" + str(test_index))
```

```python
def get_data(complete_data, 
             train_indexes, 
             test_indexes):
    train_data_in = []
    train_data_out = []
    for x in train_indexes:
        train_data = complete_data[x]
        train_data_in.append([train_data[0], train_data[1], train_data[2], train_data[3]])
        train_data_out.append(train_data[4])
    test_data_in = []
    test_data_out = []
    for x in test_indexes:
        test_data = complete_data[x]
        test_data_in.append([test_data[0], test_data[1], test_data[2], test_data[3]])
        test_data_out.append(test_data[4])
    return train_data_in, train_data_out, test_data_in, test_data_out
```

```python
number_of_clusters = 3
max_iteration = 100
```

```python
for fold_id, (train_index, test_index) in enumerate(splitter.split(data_list)):
    train_data_in, train_data_out, test_data_in, test_data_out = get_data(data_list,
                                                                          train_index,
                                                                          test_index)
    if fold_id == 0:
        print(test_data_in)
```

```python
accs = []

for fold_id, (train_index, test_index) in enumerate(splitter.split(data_list)):
    train_data_in, train_data_out, test_data_in, test_data_out = get_data(data_list,
                                                                          train_index,
                                                                          test_index)
    fcm_model = FCM(n_clusters=number_of_clusters,
                    max_iter=max_iteration)
    fcm_model.fit(train_data_in, train_data_out)
    predicted = numpy.argmax(fcm_model.predict(numpy.array(test_data_in)), 
                             axis=1)
    real = numpy.array(test_data_out)
    print(real)
    print(predicted)
    score = accuracy_score(real,
                                    predicted)
    print("Accuracy of Fold " + str(fold_id) + ": " + str(score))
    accs.append(score)
```

```python
print("Avg Accuracy: " + str(sum(accs) / len(accs)))
```
